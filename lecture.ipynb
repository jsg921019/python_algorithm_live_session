## 1. 특별한 일진법 계산기

-00+0000-0-00000

-00 +0000 -0 -00000

-1, 3, 0, -4

-2

```python
def solution(expression):
    
    l = []
    prev = None
    sign = 1
    
    for c in expression:
        
        if c == '0':
            if prev == '0':
                l[-1] += sign
            else:
                l.append(0)
        else: 
            if prev == '+' or prev == '-':
                return 'ERROR'
            else:
                if c == '+':
                    sign = 1
                else:
                    sign = -1
        
        prev = c
        
    if prev =='+' or prev == '-':
        return 'ERROR'

    return str(sum(l))
```

```python
def solution(receipt, inventory):
    inven = set(inventory)
    do_again = True
    while do_again:
        do_again = False
        for i1, i2, r in receipt:
            if i1 in inven and i2 in inven:
                inven.remove(i1)
                inven.remove(i2)
                inven.add(r)
                do_again = True
    return sorted(inven)
```

```python
from collections import defaultdict, deque

def solution(receipt, inventory):
    
    receipt = [tuple(com) for com in receipt]

    item2node = {}
    for i1, i2, r in receipt:
        item2node[i1] = (i1, i2, r)
        item2node[i2] = (i1, i2, r)
        
    graph = defaultdict(list)
    num_childs = defaultdict(int)
    for i1, i2, r in receipt:
        if r in item2node:
            graph[(i1, i2, r)].append(item2node[r])
            num_childs[item2node[r]] += 1

    order = []
    queue = deque()
    
    for com in receipt:
        if num_childs[com] == 0:
            queue.append(com)
    
    while queue:
        com = queue.popleft()
        order.append(com)
        for parent in graph[com]:
            num_childs[parent] -= 1
            if num_childs[parent] == 0:
                order.append(parent)
                
    inven = set(inventory)
    for i1, i2, r in order:
        if i1 in inven and i2 in inven:
            inven.remove(i1)
            inven.remove(i2)
            inven.add(r)
            
    return sorted(inven)
```

```python
def num_flip(n):
    cnt = 0
    for i in range(1, n+1):
        if n % i == 0:
            cnt +=1
    return cnt

def isPrime(n):
    if n == 1:
        return False
    for d in range(2, n//2+1):
        if n % d == 0:
            return False
    return True

def solution(k):
    
    n_back = 0
    n = 1
    while True:
        if num_flip(n) % 2 == 1:
            n_back += 1
            if n_back == k:
                start = n
            elif n_back == k+1:
                end = n
                break
        n += 1
        
    ans = 0
    for n in range(start, end):
        if isPrime(n):
            ans += 1
    
    return ans
```
